```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I'm excited to see someone with similar interests")
```

```python
# other pipelines available https://huggingface.co/docs/transformers/main_classes/pipelines
```

```python
# Text generation

generator = pipeline("text-generation")
generator("In this course, you will be learning about Chess.")
```

* In the above examples, no model was supplied. So HF used the default model.

* However, it is possible to load model from the HF Hub, and use it in the pipeline

```python
generator = pipeline("text-generation", model = "distilgpt2")

generator("Betty bought some bitter butter.",
          max_length = 100,
          num_return_sequences = 3)
```

```python
# Fill Mask

unmasker = pipeline("fill-mask")
unmasker("Vishwanathan Anand is an expert in playing <mask> game", top_k=3)
```

```python
unmasker("Magnus Carlsen is an expert in playing <mask>", top_k=3)
```
